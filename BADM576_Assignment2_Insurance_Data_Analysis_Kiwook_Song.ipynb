```python
import pandas as pd

# Load the Excel file
df = pd.read_csv("https://raw.githubusercontent.com/kyook17/BADM576/main/csv/insurance_marketing_dataset.csv")
print(df.shape, "\n")
df.head()
```

```python
# Select the columns for analysis
columns_for_analysis = [
    'Response',                 # Nominal, Engagement Metrics
    'Education',                # Ordinal, Customer Demographics
    'Income',                   # Continuous, Customer Demographics
    'Number of Policies',       # Discrete, Policy Information
    'Total Claim Amount',       # Continuous, Financial Information
]
```

```python
# Extract the selected columns for descriptive analysis
selected_columns_data = df[columns_for_analysis]

# Perform descriptive analysis for each selected column

# For the nominal column 'Response'
customer_response = selected_columns_data['Response'].value_counts()
print("Response Stats:\n", customer_response)
```

```python
# For the ordinal column 'Education'
education_order = ['High School or Below', 'Bachelor', 'College', 'Master', 'Doctor']
education_counts = selected_columns_data['Education'].astype('category')
education_counts = education_counts.cat.set_categories(education_order, ordered=True)
education_counts = education_counts.value_counts().sort_index()
print("Education Stats:\n", education_counts)
```

```python
# For the continuous column 'Income'
income_stats = selected_columns_data['Income'].describe()
print("Income Stats:\n", income_stats)
```

```python
# For the discrete column 'Number of Policies'
num_of_policies_stats = selected_columns_data['Number of Policies'].describe()
num_of_policies_freq = selected_columns_data['Number of Policies'].value_counts()
print("Number of Policies Stats:\n", num_of_policies_stats)
print("Number of Policies Frequencies:\n", num_of_policies_freq)
```

```python
# For the continuous column 'Total Claim Amount'
total_claim_amount_stats = selected_columns_data['Total Claim Amount'].describe()
print("Total Claim Amount Stats:\n", total_claim_amount_stats)
```
